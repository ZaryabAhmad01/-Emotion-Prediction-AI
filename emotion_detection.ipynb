```python
import pandas as pd 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score , precision_score,recall_score

from sklearn.naive_bayes import MultinomialNB,BernoulliNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
```

```python
df=pd.read_csv("I:\\NLP/emotions.csv")
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
# our data is already labeled we dont need to labelencoding 
# Your emotion label mapping:
# 0 = sadness
# 1 = joy 
# 2 = love
# 3 = anger
# 4 = fear
# 5 = surprise
```

```python
tf = TfidfVectorizer(
    max_features=5000,   # keep only top 5000 words
    min_df=2,            # ignore rare words
    max_df=0.8, # ignore very common words
    ngram_range=(1,2),
    stop_words="english"          
)
```

```python
x=tf.fit_transform(df["text"])
y=df["label"]
```

```python
x_train, x_test , y_train , y_test =train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100,model.score(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print('Accuracy score',accuracy_score(y_test,y_pred))
# print('Precision score',precision_score(y_test,y_pred))
# # print('recall  score',recall_score(y_test,y_pred))
```

```python
user_text=["This makes me so mad I could scream!"] # anger
new_user=tf.transform(user_text)
```

```python
pred_fellings=model.predict(new_user)
if(pred_fellings ==0):
    print("Sadness")
elif(pred_fellings == 3):
    print("Anger")
elif(pred_fellings == 4):
    print("Fear")
else:
    print("joy ")
```
